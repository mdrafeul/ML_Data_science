```python
##Making a list of urls from Expanded url.
#urls = data['entities'][1]['urls']
#print(urls[0]['expanded_url'])
def url_from_expanded(data):
    urls = []
    length = len(data)
    for i in range (length):
        entities = data['entities'][i]
        if entities is None:
            urls.append('NO URL')
        elif 'urls' in entities:
            length = entities['urls'][0]['expanded_url']
            url = entities['urls'][0]['expanded_url']
            urls.append(url)
        else: urls.append('NO URL')
    num_url_available=  url_count(urls)
    return(num_url_available)
```

```python
def url_count(urls): 
    count = 0 
    count_no_url = 0
    total_tweets = len(urls)
    for i in urls:
        if i == "NO URL":
            count_no_url = count_no_url + 1
        else: count = count + 1
    #print(count)
    return (count, count_no_url)
```

```python
##Find Number of Tweets for each cluster
file_names = []
total_tweets = []
total_urls = []
total_no_urls = []
for file_name in all_files:
    df = pd.read_json(file_name)
    path_to_name = os.path.basename(os.path.realpath(file_name)) ##Extract only the file name instead whole path
    file_names.append(path_to_name)
    total_tweets.append(len(df))
    num_url_available = url_from_expanded(df)
    total_urls.append(num_url_available[0])
    total_no_urls.append(num_url_available[1])
df1 = pd.DataFrame({'file_name':file_names, 'total_tweets':total_tweets,'total_urls':total_urls,'total_no_urls':total_no_urls})
df1.head(5)
```

```python
def get_orginal_url(url):
    #Define a list of shorten url services found in db
    shorten_urls = ['ow.ly','tinyurl.com','youtu.be','t.eric.sn','bit.ly','bro.in','bloom.bg','lnkd.in','cogniz.at','cs.is','dlvr.it','v.aa.com.tr','ift.tt','buff.ly','sdx.io','tex.io','hubs.li','m.eric.sn','gag.gl','httpslink.com','tirb.al','mol.im','dailym.ai','disq.us','sprou.tt','reut.rs','cnn.asia','zpr.io','invst.ly','j.mp','tmbir.co','okt.to','qoo.ly','is.gd','hw-post.com','okt.to','go.shr.lc','rb.gy','flip.it','nwslt.co','mub.me','go.ty.nl','ny.h24.news','wapo.st','clikhe.re','s68mv.app.goo.gl','trib.al','cnn.asia','trib.al','a.msn.com']
    ##This only link found problem during request
    if url == 'http://bit.ly/2O1SI0y':
        return(url)
    
    #Spiliting url
    df_temp = pd.DataFrame()
    df_temp['protocol'],df_temp['domain'],df_temp['path'],df_temp['query'],df_temp['fragment'] = zip(*[urllib.parse.urlsplit(url)])
    split_domain = df_temp['domain'].str.split(pat=".")
    length = len(split_domain[0])
    
    if len(split_domain[0][0])==1:
        orginal_url = request_url(url)
        return(orginal_url)
    elif df_temp.domain[0] in shorten_urls:
        orginal_url = request_url(url)
        return(orginal_url)
    elif split_domain[0][length-1]!=('com'or'net' or 'org' or 'edu' or 'uk'):
        #return(url)
        orginal_url = request_url(url)
        return(orginal_url)
    else: return(url)
    
        
```

```python
def request_url(url):
    try:
        r = requests.get(url,timeout=10)
        orginal_url = r.url
        return(orginal_url)
    except: return(url) 
```

### Exploring Domain Names

```python
#read all file from the directory
path = r'C:/Users/rafe7/thesis_mainclaims/Twitter Data/tweets_url' 
all_files_urls = glob.glob(path + "/*.csv")
```

```python
df_domain = pd.DataFrame()
get_domain = []
for file_name in all_files_urls:
    df_domain_current_file = []
    path_to_name = os.path.basename(os.path.realpath(file_name))
    if path_to_name != 'all_tweets_urls.csv':
        df_temp = pd.read_csv(file_name)
        for i in df_temp['urls']:
            ##Getting the domain name from url
            domain = urlparse(i).netloc
            get_domain.append(domain)        
##createing dataframe
df_domain = pd.DataFrame({'domain':get_domain})

##group by domain name and delete duplicates urls 
df_domain['count'] = df_domain.groupby('domain')['domain'].transform('count') 
df_domain = df_domain.drop_duplicates(subset=["domain"])
df_domain.reset_index(drop=True, inplace=True)
```
